#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv("assets/real_estate.csv", sep=";")
south_belt = ds[ds["level5"].isin(["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"])]
south_belt["normalized_price"] = south_belt.groupby("level5")["price"].transform(lambda x: (x - x.mean()) / x.std())


fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

subset_fuenlabrada = south_belt[south_belt["level5"] == "Fuenlabrada"]["normalized_price"]
axes[0, 0].hist(subset_fuenlabrada, bins=20, alpha=0.7, color='blue')
axes[0, 0].set_title("Fuenlabrada")
axes[0, 0].set_xlabel("Normalized Price")
axes[0, 0].set_ylabel("Frequency")

subset_leganes = south_belt[south_belt["level5"] == "Leganés"]["normalized_price"]
axes[0, 1].hist(subset_leganes, bins=20, alpha=0.7, color='green')
axes[0, 1].set_title("Leganés")
axes[0, 1].set_xlabel("Normalized Price")
axes[0, 1].set_ylabel("Frequency")

subset_getafe = south_belt[south_belt["level5"] == "Getafe"]["normalized_price"]
axes[1, 0].hist(subset_getafe, bins=20, alpha=0.7, color='red')
axes[1, 0].set_title("Getafe")
axes[1, 0].set_xlabel("Normalized Price")
axes[1, 0].set_ylabel("Frequency")

subset_alcorcon = south_belt[south_belt["level5"] == "Alcorcón"]["normalized_price"]
axes[1, 1].hist(subset_alcorcon, bins=20, alpha=0.7, color='purple')
axes[1, 1].set_title("Alcorcón")
axes[1, 1].set_xlabel("Normalized Price")
axes[1, 1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()

```

The histograms for Leganes and Fuenlabrada indicates that the average prices in these areas are generally lower than the average but not significantly so. The histogram for Alcoron shows a more pronounced tendency towards lower prices in this area compared to the overall average. The histogram for Getafe shows peaks near both 1 and -1. This distribution indicates that there are two dominant price groups both way above and below the average.

#### Exercise 21. Make a plot of the coordinates (latitude and longitude columns) of the south belt of Madrid by color of each population (you must use the subset obtained in Exercise 14) (★★★★)

Execute the following cell, and then start coding in the next one. You must implement a simple code that transforms the coordinates columns in a Python dictionary (add more information if needed) and then add it to the map

```python
from ipyleaflet import Map, basemaps,  CircleMarker, Polyline

# Mapa centrado en (60 grados latitud y -2.2 grados longitud)
# Latitud, longitud
center = [60, -2.2]
zoom = 2
map = Map(center = center, zoom = zoom, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.Esri.WorldTopoMap)
```

```python

import pandas as pd
import matplotlib.pyplot as plt
ds = pd.read_csv("assets/real_estate.csv", sep=";")
mad_south = ds[ds["level5"].isin(["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"])]
center = [40.31 , -3.7]
zoom =  12
map = Map(center = center, zoom = zoom, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.Esri.WorldTopoMap)

# We check there are no NA values or 0 values.
mad_south = mad_south.dropna(subset=['latitude', 'longitude'])
mad_south = mad_south[(mad_south['latitude'] != 0) & (mad_south['longitude'] != 0)]

# We substitute the "," for ".", and avoid errors for running the same code again.
if mad_south['latitude'].dtype == 'object':
    mad_south['latitude'] = mad_south['latitude'].str.replace(',', '.').astype(float, errors='ignore')

if mad_south['longitude'].dtype == 'object':
    mad_south['longitude'] = mad_south['longitude'].str.replace(',', '.').astype(float, errors='ignore')

# We create the dictionary to introduce the coordinates.
coordinates = {}

# We create a loop to add the data into the dicctionary, keeping in mind we will have multiple values for the same key.
for index, row in mad_south.iterrows():
    poblacion = row['level5']
    latitude = row['latitude']
    longitude = row['longitude']
    if poblacion not in coordinates:
        coordinates[poblacion] = []
    coordinates[poblacion].append((latitude, longitude))

# Print the dictionary content to confirm it was added correctly.
print(coordinates)

# Color dictionary for each poblacion.
color_dict = {
    'Fuenlabrada': 'blue',
    'Leganés': 'green',
    'Getafe': 'red',
    'Alcorcón': 'purple'
}

# We add circles on each location, with the colour of its respective poblacion. If there is data with different information from our color_dict, it will display it in black.
for poblacion, coords_list in coordinates.items():
    color = color_dict.get(poblacion, 'black')
    for coords in coords_list:
        circle_marker = CircleMarker(
            location=coords,
            radius=5,  
            color=color,
            fill=True,
            fill_color=color,
            fill_opacity=0.6,
            title=poblacion
        )
        map.add_layer(circle_marker)

    # We connect the dots for their locations with the same color.
    polyline = Polyline(
        locations=coords_list,
        color=color,  
        weight=1,
        opacity=1
    )
    map.add_layer(polyline)

# We display the map.
map
```
